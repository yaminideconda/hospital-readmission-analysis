```python
import pandas as pd
import os

# Define the file path (use forward slashes!)
file_path = "C:/Users/yamin/OneDrive/Documents/hospital-readmission-analysis/Data/diabetic_data.csv"

# Check if the file exists
if os.path.exists(file_path):
    # Load the dataset
    df = pd.read_csv(file_path)
    
    print("‚úÖ Dataset loaded successfully!")
    print("Shape of dataset:", df.shape)
    
    # Display first few rows
    display(df.head())
else:
    print("‚ùå File not found at:", file_path)
```

```python
# Basic info
print("üßæ Dataset Info:")
df.info()

# Column names
print("\nüß© Column Names:")
print(df.columns.tolist())

# Data types
print("\nüîç Data Types:")
print(df.dtypes)

# Summary statistics
print("\nüìä Summary Stats (Numerical Columns):")
display(df.describe())

# Check for missing values
print("\nüßº Missing Values (per column):")
print(df.isnull().sum())
```

```python
# Replace '?' with pandas' NA value
df.replace('?', pd.NA, inplace=True)
```

```python
# Replace '?' with pandas' NA value
df.replace('?', pd.NA, inplace=True)
```

```python
# Drop columns with excessive missing data
df.drop(columns=['weight', 'payer_code', 'medical_specialty'], inplace=True)
```

```python
# Drop identifier columns
df.drop(columns=['encounter_id', 'patient_nbr'], inplace=True)
```

```python
# Count values in 'readmitted'
df['readmitted'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='readmitted')
plt.title("Readmission Distribution")
plt.xlabel("Readmitted")
plt.ylabel("Number of Patients")
plt.show()
```

```python
# Compare readmission across genders
sns.countplot(data=df, x='gender', hue='readmitted')
plt.title("Readmission by Gender")
plt.show()
```

```python
# Age group vs readmission
sns.countplot(data=df, x='age', hue='readmitted')
plt.title("Readmission by Age Group")
plt.xticks(rotation=45)
plt.show()
```

```python
df['time_in_hospital'].describe()
```

```python
sns.histplot(df['time_in_hospital'], bins=10, kde=True)
plt.title("Time in Hospital Distribution")
plt.xlabel("Days")
plt.ylabel("Number of Patients")
plt.show()
```

```python
# Only include numeric columns
numeric_df = df.select_dtypes(include='number')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Numerical Features")
plt.show() 
```
